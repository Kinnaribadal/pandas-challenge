```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(purchase_data_file)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#we are creating a list without duplicates
player_count = purchase_data.loc[:,["Gender", "SN", "Age"]].drop_duplicates()
#getting count from first row
num_players = player_count.count()[0]
#for display purposes we are puting it in dataframe
pd.DataFrame({"Total Players": [num_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#getting unique items with unique fuction and assigining to a dataframe
unique = purchase_data["Item Name"].unique()
unique_df = pd.DataFrame(unique)
#counting number of unique items
count = unique_df.count()
count
#finding avereage and rounding to 2 digits
average = round(purchase_data["Price"].mean(),2)
average
#finding the total and rounding to 2 digits
total = round(purchase_data["Price"].sum(),2)
total
#
#displaying results by assigning to dataframe
summary_dicts = {"Unique Items": count, "Average":"$" + str(average), "Purchases":num_players, "Total":"$" + str(total)}
summary_df = pd.DataFrame(summary_dicts)
summary_df
      
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#counting total number of players by different value in the list
gender_count = player_count["Gender"].value_counts()
gender_count
#calculating percentage num_player we created earlier
gender_percent = (gender_count/num_players)*100
gender_percent
#displaying results by assigning to dataframe using .map function for formating
gender_df = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": gender_percent})
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2f}%".format)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#calculating purchase total per gender next count per gender and average per gender
gender_total = purchase_data.groupby(["Gender"]).sum()["Price"]
gender_count = purchase_data.groupby(["Gender"]).count()["Price"]
gender_average = purchase_data.groupby(["Gender"]).mean()["Price"]
purchase_per_person = gender_total/gender_df["Total Count"]
purchase_analysis_df = pd.DataFrame({"Purchase Count":gender_count,
                                     "Average Purchase Price": gender_average,
                                     "Total Purchase Value": gender_total,
                                     "Avg Total Purchase per Person": purchase_per_person})
#displaying results by assigning to dataframe
purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_df["Avg Total Purchase per Person"] = purchase_analysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#creating bins
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
player_count=purchase_data
player_count["Age Ranges"]=pd.cut(player_count["Age"], age_bins, labels = group_names)
player_count=player_count.groupby("Age Ranges")
player_count.count()
summary_df = pd.DataFrame(player_count.count())
summary_df


summary_df["Purchase ID"] = (summary_df["Purchase ID"]/num_players)*100
summary_df 

# format percentages 
summary_df["Purchase ID"] = summary_df["Purchase ID"].map("{:,.2f}%".format)
summary_df

#  Reduce number of columns 
age_df = summary_df[["Purchase ID","SN"]]
age_df

# Rename the columns for the final age dempgraphic df using .rename(columns={})
final_summary_df = age_df.rename(columns={"Purchase ID":"Percentage of Players", "SN":"Total Count"})
final_summary_df

```
